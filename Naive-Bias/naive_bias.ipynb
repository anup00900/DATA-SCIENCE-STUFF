```python
class NaiveBias:
  def __init__(self):
    pass

  def fit(self,X,Y):
    self.prob={}
    self.dummy_label = 'Yes'
    self.features = X.columns
    self.label_prob = Y.value_counts(normalize=True)
    for col in X.columns:
      self.prob[col]=pd.crosstab(X[col],Y,normalize = 'columns')
      #As we can see columns are here of text type.
    return self.prob  

  def pred(self,x):
    maxprob=0
    maxlab=""
    for lab in self.label_prob.index:
      currprob = self.label_prob[lab]
      currlab=lab

      for fet in self.features:
        p = self.prob[fet][lab][x[fet]]
        currprob = currprob*p

    
      if currprob > maxprob:
        maxprob = currprob
        maxlab = currlab
    return maxlab  





  def predict(self,X):
    y_pred = np.full(X.shape[0],self.dummy_label)
    for i,(_,x) in enumerate(X.iterrows()): #Har ek sample (instances Ke liye) usko uska column name + uska value Pass kr rhe not sirf value paas kr rhe
      print(i,x)
      lab = self.pred(x)
      y_pred[i]=lab
    
    return y_pred    


    
    
```

```python
model = NaiveBias()
prob = model.fit(X,Y)


#Jaise Yaha pr X mein categorical data hai khud se likha hua code pr train ho ja rha hai
#agr sklearn pr fit krwaye toh nhi hoga ek drwaback of sklearn
```
