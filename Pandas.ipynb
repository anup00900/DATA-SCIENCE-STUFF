# Concatinate Data Frame

Creating Data Frame

```python
india_weather = pd.DataFrame({"city":["mumbai","delhi","banglore"],"temperature":[32,45,30],"humidity":[80,60,78]})
india_weather
```

```python
us_weather = pd.DataFrame({"city":["chicago","new_york","orlando"],"temperature":[21,14,35],"humidity":[68,65,75]})
us_weather
```

```python
df1 = pd.concat([us_weather,india_weather],ignore_index = True)
df1
```

Concatenating Data Frame Column Wise

```python
df = pd.concat([india_weather,us_weather],axis = 1)
df
```

**Merge Data Frames**

```python
us_temp = pd.DataFrame({"city":["chicago","new_york","orlando"],"temperature":[21,14,35]})
us_temp
```

```python
us_hum= pd.DataFrame({"city":["chicago","new_york"],"humidity":[68,65]})
us_hum
```

```python
df2 = pd.merge(us_temp,us_hum,on='city',how='outer')
df2
```
